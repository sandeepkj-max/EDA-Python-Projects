```python
def ensure_date_format(df, col, fmt):
        df[col] = pd.to_datetime(df[col], errors='coerce')  # flexible parsing
        df[col] = df[col].dt.strftime(fmt)  # convert to required final format
        return df
```

```python
df = ensure_date_format(df, "reservation_status_date", "%Y-%m-%d")
```

```python
###Average Daily Rate trend over three years
```

```python

def get_month(x):
  pre = ''
  if Months.index(x)<9:
    pre = '0'
  return pre+str(Months.index(x)+1)

def get_day(x):
  pre = ''
  if x<10:
    pre = '0'
  return pre+str(x)
```

```python
df2.describe()
```

```python

df2['arrival_date'] = df2.arrival_date_year.apply(lambda x: str(x))+" - "+df2.arrival_date_month.apply(get_month)+"-"+df2.arrival_date_day_of_month.apply(get_day)
df2.head(2)
     
```

```python
d = df2.groupby(['hotel','arrival_date'])['adr'].mean().reset_index().sort_values('arrival_date')
# fig = plt.figure(figsize=(20,7))
fig, ax = plt.subplots(figsize=(20, 7))
sns.lineplot(x='arrival_date', y='adr', hue='hotel', data=d)
plt.xlabel("Date")
plt.ylabel("Average Daily Price")
plt.grid()
# fig.autofmt_xdate()
p = plt.xticks(rotation=30)
ax.tick_params(axis='x', labelsize=3)
plt.title("Average daily rate trend over three years")
```

```python
## Here we observe that Highest daily price (for each year) is increasing every year.
```

```python
###CANCELLATIONS
```

```python
df.isna().sum()
```

```python
d = df.groupby(['hotel','arrival_date_month']).agg({'is_canceled':'sum', 'meal':'count'}).reset_index().rename(columns={'is_canceled':'cancelled_bookings', 'meal':'total_bookings'})
d['cancellation_percentage'] = (d['cancelled_bookings']/d['total_bookings'])*100
d.head()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='arrival_date_month', y='cancellation_percentage', hue='hotel', data=d, order=Months)
plt.ylim((0,100))
plt.xticks(rotation=30)
plt.xlabel("Months")
plt.ylabel("% Cancellation")
plt.title("Cancellations per month")
```

```python
#Effect of Lead time on Cancellations
```

```python
d = df.groupby('lead_time').agg({'is_canceled':'sum', 'hotel':'count'}).reset_index().rename(columns={'is_canceled':'cancelled_bookings', 'hotel':'total_bookings'})
d['cancellation_percentage'] = (d['cancelled_bookings']/d['total_bookings'])*100
d.head()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='lead_time', y='cancellation_percentage', data=d)
plt.ylim((0,100))
plt.xlabel("Lead Time (days)")
plt.ylabel("% Cancellation")
plt.title("Effect of Lead time on Cancellations")
```
